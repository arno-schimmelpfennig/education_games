### But what if you want to use this package as a CLI tool in your terminal?

In a lot of cases it is useful to use your python package as a CLI tool in your terminal that you can use without importing it into a python file and add the inputs in the terminal.

For that we need to make a python module that accepts also inputs. So let's have a look into the ```main.py```. In that file we see one way of how we can create cli tools: the python library ```typer```. There are also other ways like ```click``` or ```argparse``` but typer offers a nice and easy way to also define defaults and help outputs. All of these libraries transform your python package into a cli tool and make it possible to add inputs in the terminal. 

So let's say we want to run the ```add_one``` function from the ```example_file.py``` in the terminal and add an integer as input that one is added to.

In the ```main.py``` we have to import the ```add_one``` module from the package and of course ```typer``` itself:

```python
from example.example_file import add_one
import typer
```

Then we have to call typer:

```python
app = typer.Typer()
```

This will create a new typer app, which we can use to add commands to and call in the terminal.

And then we can create a function that calls the function that we want to run as a cli app. But we have to use the ```decorator: @app.command()``` (see below what a decorator does) with this function. Additionally you can define the input you want to hand over in the terminal in the function input ```x: int = typer.Option(0, help='enter an integer')```. The ```x``` is the name of the input and the ```int``` is the type of the input. For example if we run: ```python -m example.main run_add_one --x 5``` the output in the terminal will be 6. The ```help``` is the text that is shown when you call ```python -m example.main --help```.

```python
@app.command(help="adds one to the input")
def run_add_one(x: int = typer.Option(0, help='enter an integer')):
    print(add_one(x))
```

Now the last step is to start the app if you call this main.py file directly. You can do that with the if-statement:

```python
if __name__ == "__main__":
    app()
```

Before executing code, the Python interpreter reads the module and defines few special variables/global variables. 
If the python interpreter is running that module as the main program, it sets the special __name__ variable to have a value “__main__”. If this file is being imported from another module, __name__ will be set to the module’s name. Module’s name is available as value to __name__ global variable. 

So the app will only start if ```main.py``` is invoked directly and not imported to another file.

If you change something in the files run ```pip install src/.```.
Now you can run the python package as cli tool in your terminal using:

```sh
python -m example.main --help
```


CLI Tools can be helpful for example if you want to train a model or get a prediction from a model from the terminal and not in your juptyer notebook.

#### Interlude: Decorators

In this section you will find python decorators. A decorator is a design pattern in Python that allows a user to add new functionality to an existing object without modifying its structure. Decorators are usually called before the definition of a function you want to decorate.

Example:
We will create a simple function that will convert a string to uppercase. For it to work as a decorator we need to create a function that includes a second wrapper function.

````python
def uppercase_decorator(function):
    def wrapper():
        func = function()
        make_uppercase = func.upper()
        return make_uppercase

    return wrapper
````

Our decorator function takes a function as an argument, and we shall, therefore, define a function and pass it to our decorator.

````python
def say_hi():
    return 'hello there'

decorate = uppercase_decorator(say_hi)
decorate()
````

However, Python provides a much easier way for us to apply decorators. We simply use the ```@ symbol``` before the function we'd like to decorate. 

````python
@uppercase_decorator
def say_hi():
    return 'hello there'

say_hi()
````
