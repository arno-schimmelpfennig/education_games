# Intro to Python Packaging

Python packaging is used to structure your python code in a way that it can be easily reused by others and yourself in other projects. It is also a way to make your code more readable and maintainable. In most cases it is an iterative process and you will learn more about it as you go and use python. One example when you could use python packaging is when you want to create a python package that contains all your data science helper functions. This way you can easily import them in other projects and you don't have to copy and paste them all the time. For example a plotting function. You would start writing the function in your juptyer notebook and once you are happy with it you can move it to a python package. This way you can use it in other notebooks and projects and make the notebook more readable, because you have less code in it.

For this notebook we highly recommend to split the vs code editor you can do that by drag and drop this notebook or the files mentioned in this notebook to the right or press on the split editor button in the upper right corner.


In this repository you will find a ```src``` directory and in this directory there are several folders. In this notebook we only care about the ```example``` one.

<pre>
ds-unit-testing
├──src/
│  ├──example/
│  │  ├── __init__.py
│  │  ├── example_file.py
│  │  └── main.py
</pre>

In this example directory you will find two python files:

```__init__.py``` is required to import the directory as a package, and should be empty.

```example_file.py``` is an example of a module within the package that could contain the logic (functions, classes, constants, etc.) of your package.

```main.py``` is an example of a module within the package that calls other moduls and accepts inputs.

Normally the directory containing the Python files should match the project name you choose for your python package. This simplifies the configuration and is more obvious to users who install the package.

Packages are a way of structuring Python’s module namespace by using “dotted module names”. For example, the module name A.B designates a submodule named B in a package named A.

So in this case: ```src.example```

When importing the package, Python searches through the directories on sys.path looking for the package subdirectory. The ```__init__.py``` files are required to make Python treat directories containing the file as packages.


In the ```example_file.py``` file you will find the following function:


```python
def add_one(number):
    return number + 1
```

Because this notebook is located in the parent directory you can simply import the package like:

````python
from src.example.example_file import add_one
````

And we can use the function now in the notebook.

````python
print(add_one(1))
````

If you put your notebooks in a future project into a notebook directory and your repository looks something like:

And you want to import the preprocessing and visualization packages into your notebooks than the pyhton will still look in the sys.path for the packages. But because the path is now ```/notebooks``` it will not find them unless you tell the system where to look with:

```python
import sys
# setting path
sys.path.append('../')
from src.preprocessing.impute import *
```




You can import python files from your local directory as described above. But you can even go a step further and package this project into a real python package that you can install with ```pip```. 
You may have noticed the ```setup.py``` file in in the ```src``` directory. This file is one way to create a python package. In this file you define the name of the package and other metadata that you want to add. So have a look into that file. Now you can simply run ```pip install src/.``` in your terminal (in our case the ```setup.py``` is located in the ```src``` directory so we have to include it in the path).


The other way you can see under [this link](https://packaging.python.org/en/latest/tutorials/packaging-projects/).

After you ran ```pip install src/.``` you simply can import the package without adding ```src.``` and this part to the import:
```python
import sys
# setting path
sys.path.append('../')
```

````python
%pip install src/.
````

You can also see it if you use ```pip freeze``` in your terminal. You should see the package you just installed.

````python
!pip freeze | grep example
````

And now we can simply import the package like the other packages we installed with pip.

````python
from example.example_file import add_one

print(add_one(3))
````

At this point you could even create a source distribution with:
```sh
cd src
python setup.py sdist
```
This will create the source distribution in the ```src/dist/``` directory. You
can share this file with others or [publish it on PyPI](https://python-packaging.readthedocs.io/en/latest/minimal.html#publishing-on-pypi). 

However the main advantage is in our case, that you can package functions that you use in your notebook often into python files which you than can easily import.
