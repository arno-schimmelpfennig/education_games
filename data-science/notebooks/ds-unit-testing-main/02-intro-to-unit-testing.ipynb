# Intro to Unit Testing

Unit testing is a method for testing software that looks at the smallest testable pieces of code, called units, which are tested for correct operation. By doing unit testing, we can verify that each part of the code, including helper functions that may not be exposed to the user, works correctly and as intended. Of course this is best practice and not always possible or feasible to do. In most cases this is only enforced if you are working in a team and you have to follow certain guidelines. But it is very good to know how to do it and what the benefits are.

The idea is that, we independently check each small piece of our program to ensure that it works. It is a major part of test-driven development (TDD). In layman’s terms, TDD recommends writing tests that would check the functionality of your code prior to your writing the actual code. And thus one of the advantages of TDD is that it helps you plan out the work ahead. 

Following this process ensures that you careful plan the code you write in order to pass these tests. This also prevents the possibility of writing tests being postponed to a later date, as they might not be deemed as necessary compared to additional features that could be created during that time.

Tests also give you confidence when you begin to refactor code, as you are more likely to catch bugs due to the instant feedback when tests are executed.

![](images/test-driven-development.png)

1. Red Phase

    In this phase, based on a requirement, a user story or anything that describes the functionality, we write the test or tests to validate this functionality. It’s called the Red phase because red is the error colour used for failed tests. We always start with a failed test.

2. Green Phase

    Once we have the failed test, we implement the simplest code that will make the failed test pass. This doesn’t need to be the most elegant or efficient code. The goal is to reach a safe and stable point by moving forward in small steps. This is where we follow the KISS principle: Keep it Simple, Stupid.

3. Refactor Phase

    Once we have written a series of tests, or just one test, we move on to the Refactor phase in which we improve the code without changing the functionality. A common mistake in this phase is changing the functionality instead of improving the design. For this phase, we strongly recommend following typical refactoring patterns.

By the end of this phase, we have clean, simple and efficient code.

In this Notebook you will see how to run tests on your code and why using print statements is not the most efficient way.

But first let's define a function that we want to test. Normally in TDD you would make it the other way around write your tests first and than the functions!

````python
def divide(x, y):
    """ division function 
    >>> 3 // 2
    1.5
    >>> 6 / 2
    3
    """
    if y == 0:
        return "Can not divide by zero"
    return x // y
````

## Print()

You will get in a lot of situation where you will have to debug code and in a lot of cases you will use print-statements to have a look if the code you wrote is actually doing what you think it should be doing. But as you see below for testing your code print-statements are not really useful. If you have a unexpected result you don't see where in your code the mistake is and you have to write a lot of them to look at edge cases.

````python
print('6 / 2 Expected result: 3 Function result: ', divide(6,2))
print('5 / 5 Expected result: 1 Function result: ', divide(5,5))
print('-10 / 2 Expected result: -5 Function result: ', divide(-10,2))
print('3 / 2 Expected result: 1.5 Function result: ', divide(3,2))
````

## Pytest

Pytest is the most common library that is used in unit testing. It allows you to easily test your code with little code, but it can scale to support complex functional testing for applications and libraries too.

Pytest uses **asserts statements**. Assertions let you write sanity checks in your code. These checks are known as assertions, and you can use them to test if certain assumptions remain true while you’re developing your code. If any of your assertions turn false, then you have a bug in your code. So you can use assertions as a convenient tool for documenting, debugging, and testing code during development. So you can also use them independently from unit testing.

Example:

````python
assert 3 + 3 == 6 
assert "HALLO".lower() == "hallo"
````

Now with pytest you can use assertions for testing your code in test functions. You can add as many assertions as you need e.g. for edge cases. Edge cases are typically cases where your code might fail. 

````python
import pytest
# Because pytest doesn't run natively in Jupyter Notebooks we have to import a workaround
import ipytest
ipytest.autoconfig()
````

````python
%%ipytest -qq

def test_divide():
    assert divide(3,2) == 1.5
    assert divide(5,5) == 1
    assert divide(6,2) == 3
    assert divide(-2,0) == "Can not divide by zero"
    assert divide(10,-2) == -5
````

In every step we see that the divide functions fails when getting a decimal number as result. You may have notice that we use floor devision in the function, but because we expect the division function to return a decimal number we should change it to a normal devision and run the tests again.

### Fixtures

pytest fixtures are a way of providing data, test doubles, or state setup to your tests. Fixtures are functions that can return a wide range of values. Each test that depends on a fixture must explicitly accept that fixture as an argument.

#### When to Create Fixtures
In this section, you’ll simulate a typical test-driven development workflow.

Imagine you’re writing a function, format_data_for_display(), to process the data returned by an API endpoint. The data represents a list of people, each with a given name, family name, and job title. The function should output a list of strings that include each person’s full name (their given_name followed by their family_name), a colon, and their title:

````python
people = [
    {
        "given_name": "Alfonsa",
        "family_name": "Ruiz",
        "title": "Senior Software Engineer",
    },
    {
        "given_name": "Sayid",
        "family_name": "Khan",
        "title": "Project Manager",
    },
]
````

So let's first write the test for the expected output:

````python
%%ipytest -qq

def test_format_data_for_display():
    people = [
        {
            "given_name": "Alfonsa",
            "family_name": "Ruiz",
            "title": "Senior Software Engineer",
        },
        {
            "given_name": "Sayid",
            "family_name": "Khan",
            "title": "Project Manager",
        },
    ]

    assert format_data_for_display(people) == [
        "Alfonsa Ruiz: Senior Software Engineer",
        "Sayid Khan: Project Manager",
    ]
````

It fails. Now let's write the function.

````python
def format_data_for_display(people):
    # Implement this!
    return [f'{item["given_name"]} {item["family_name"]}: {item["title"]}' for item in people]
````

And run the test again!

````python
%%ipytest -qq

def test_format_data_for_display():
    people = [
        {
            "given_name": "Alfonsa",
            "family_name": "Ruiz",
            "title": "Senior Software Engineer",
        },
        {
            "given_name": "Sayid",
            "family_name": "Khan",
            "title": "Project Manager",
        },
    ]

    assert format_data_for_display(people) == [
        "Alfonsa Ruiz: Senior Software Engineer",
        "Sayid Khan: Project Manager",
    ]
````

While writing this test, it occurs to you that you may need to write another function to transform the data into a pandas Dataframe to possibly export it as csv:

````python
def format_data_for_excel(people):
     # Implement this!
     csv_string = "given,family,title\n"
     people_str = ''.join([f'{item["given_name"]},{item["family_name"]},{item["title"]}\n' for item in people])
     csv_string += people_str     
     return csv_string
````

 The test for the format_data_for_excel() function would look awfully similar to the format_data_for_display() function:

````python
%%ipytest -qq

def test_format_data_for_excel():
    people = [
        {
            "given_name": "Alfonsa",
            "family_name": "Ruiz",
            "title": "Senior Software Engineer",
        },
        {
            "given_name": "Sayid",
            "family_name": "Khan",
            "title": "Project Manager",
        },
    ]

    assert format_data_for_excel(people) == """given,family,title
Alfonsa,Ruiz,Senior Software Engineer
Sayid,Khan,Project Manager
"""
````

Notably, both the tests have to repeat the definition of the people variable, which is quite a few lines of code.

If you find yourself writing several tests that all make use of the same underlying test data, then a fixture may be in your future. You can pull the repeated data into a single function decorated with @pytest.fixture to indicate that the function is a pytest fixture:

````python
@pytest.fixture
def example_people_data():
    return [
        {
            "given_name": "Alfonsa",
            "family_name": "Ruiz",
            "title": "Senior Software Engineer",
        },
        {
            "given_name": "Sayid",
            "family_name": "Khan",
            "title": "Project Manager",
        },
    ]
````

````python
%%ipytest -qq

def test_format_data_for_display(example_people_data):
    assert format_data_for_display(example_people_data) == [
        "Alfonsa Ruiz: Senior Software Engineer",
        "Sayid Khan: Project Manager",
    ]

def test_format_data_for_excel(example_people_data):
    assert format_data_for_excel(example_people_data) == """given,family,title
Alfonsa,Ruiz,Senior Software Engineer
Sayid,Khan,Project Manager
"""
````

Fixtures are great for extracting data or objects that you use across multiple tests. However, they aren’t always as good for tests that require slight variations in the data. Littering your test suite with fixtures is no better than littering it with plain data or objects. It might even be worse because of the added layer of indirection.



### Parametrization

Imagine you’ve written a function to tell if a string is a palindrome. 

````python
import re
def is_palindrome(s):
    s = s.lower()
    s = s.replace(' ', '')
    s = re.sub('[^A-Za-z0-9]+', '', s)
    return s == s[::-1]
````

An initial set of tests could look like this:

````python
%%ipytest -qq

def test_is_palindrome_empty_string():
    assert is_palindrome("")

def test_is_palindrome_single_character():
    assert is_palindrome("a")

def test_is_palindrome_mixed_casing():
    assert is_palindrome("Bob")

def test_is_palindrome_with_spaces():
    assert is_palindrome("Never odd or even")

def test_is_palindrome_with_punctuation():
    assert is_palindrome("Do geese see God?")

def test_is_palindrome_not_palindrome():
    assert not is_palindrome("abc")

def test_is_palindrome_not_quite():
    assert not is_palindrome("abab")
````

All of these tests except the last two have the same shape:
```python
def test_is_palindrome_<in some situation>():
    assert is_palindrome("<some string>")
```

This is starting to smell a lot like boilerplate. pytest so far has helped you get rid of boilerplate, and it’s not about to let you down now. You can use @pytest.mark.parametrize() to fill in this shape with different values, reducing your test code significantly:

````python
%%ipytest -qq

@pytest.mark.parametrize("palindrome", [
    "",
    "a",
    "Bob",
    "Never odd or even",
    "Do geese see God?",
])
def test_is_palindrome(palindrome):
    assert is_palindrome(palindrome)

@pytest.mark.parametrize("non_palindrome", [
    "abc",
    "abab",
])
def test_is_palindrome_not_palindrome(non_palindrome):
    assert not is_palindrome(non_palindrome)
````

The first argument to parametrize() is a comma-delimited string of parameter names. You don’t have to provide more than one name, as you can see in this example. The second argument is a list of either tuples or single values that represent the parameter value(s). You could take your parametrization a step further to combine all your tests into one:

````python
%%ipytest -qq

@pytest.mark.parametrize("maybe_palindrome, expected_result", [
    ("", True),
    ("a", True),
    ("Bob", True),
    ("Never odd or even", True),
    ("Do geese see God?", True),
    ("abc", False),
    ("abab", False),
])
def test_is_palindrome(maybe_palindrome, expected_result):
    assert is_palindrome(maybe_palindrome) == expected_result
````
